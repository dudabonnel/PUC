```python
import numpy as np
import matplotlib.pyplot as plt
import random

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

### Exemplo 1

```python
X, y = make_blobs(n_samples = 1000, centers=[[-1,-1], [1,1]], cluster_std = 0.25)
```

```python
plt.plot(X[:,0], X[:,1], '.', color='gray')
#plt.plot(X[:,0], X[:,1], '.', color='gray', alpha = 0.5)
#plt.plot(X[:,0],X[:,1],'.',marker='o', markerfacecolor='gray', color='black', alpha=0.5)
```

```python
k_means = KMeans(n_clusters = 2)
k_means.fit(X)
```

```python
k_means.labels_
```

```python
k_means.cluster_centers_
```

```python
Xc = k_means.cluster_centers_
tags = k_means.labels_

plt.plot(X[ tags==0 ,0], X[ tags==0 ,1], '.', color = 'blue')
plt.plot(X[ tags==1 ,0], X[ tags==1 ,1], '.', color = 'yellow')
plt.plot( Xc[:,0], Xc[:,1], 'ro')    
```

### Exemplo 2

```python
np.random.seed(123)
```

```python
X, y = make_blobs(n_samples = 5000, centers=[[4,4], [-2,-1], [2,-3], [1,1]], cluster_std = 0.9)
```

```python
plt.plot(X[:,0], X[:,1], '.', color='gray')
```

```python
k_means = KMeans(n_clusters = 4, n_init = 12)
k_means.fit(X)
```

```python
k_means.labels_
```

```python
Xc = k_means.cluster_centers_
tags = k_means.labels_
cores = ['blue', 'yellow', 'magenta', 'green']

for tag, cor in zip( np.unique(tags), cores ):
    plt.plot(X[ tags==tag ,0], X[ tags==tag ,1], '.', color = cor, alpha=0.5)
plt.plot( Xc[:,0], Xc[:,1], 'ro')     
```

```python
# k =3
k_means = KMeans(n_clusters = 3, n_init = 12)
k_means.fit(X)

Xc = k_means.cluster_centers_
tags = k_means.labels_
cores = ['blue', 'yellow', 'magenta', 'green']

for tag, cor in zip( np.unique(tags), cores ):
    plt.plot(X[ tags==tag ,0], X[ tags==tag ,1], '.', color = cor, alpha=0.5)
plt.plot( Xc[:,0], Xc[:,1], 'ro')  
```

```python
# k =2
k_means = KMeans(n_clusters = 2, n_init = 12)
k_means.fit(X)

Xc = k_means.cluster_centers_
tags = k_means.labels_
cores = ['blue', 'yellow', 'magenta', 'green']

for tag, cor in zip( np.unique(tags), cores ):
    plt.plot(X[ tags==tag ,0], X[ tags==tag ,1], '.', color = cor, alpha=0.5)
plt.plot( Xc[:,0], Xc[:,1], 'ro')  
```

### Determinação da quantidade de Clusters

```python
TWSS = []
K = []
for k in range(2,11):
    k_means = KMeans(n_clusters = k, n_init = 12)
    k_means.fit(X)
    TWSS.append( k_means.inertia_ )
    K.append(k)
```

```python
plt.plot(K, TWSS, 'ko--')
plt.xlabel('k')
plt.ylabel('Total Within Sum of Square')
plt.plot([4,4], [min(TWSS),max(TWSS)], 'k')
```
