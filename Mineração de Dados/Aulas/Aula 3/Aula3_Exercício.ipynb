# Analisando dados de segurança pública

```python
import pandas as pd
```

## Carregando os dados

```python
df = pd.read_csv('Furto_celular_2022.csv', sep=';', low_memory=False)
df.head()
```

## Filtrando somente a capital

```python
df[df.CIDADE == 'S.PAULO']
```

```python
#df[df.CIDADE == 'S.PAULO']
df[ df['CIDADE'] == 'S.PAULO' ]
```

## Agregando por bairro e ordenando

```python
df[df.CIDADE == 'S.PAULO'].groupby('BAIRRO')['NUMERO_BOLETIM'].count().sort_values(ascending=False).to_frame()
```

## Contagem de furtos por marca do aparelho

```python
df[df.CIDADE=='S.PAULO'].groupby('MARCA_CELULAR').agg({'NUMERO_BOLETIM':'count'}).sort_values(by='NUMERO_BOLETIM', ascending=False)
```

## Exemplo de como incluir escritas distintas

```python
lista = ['RUA MONTE ALEGRE', 
         'RUA MONTE ALEGRE, 984',
         'Rua Monte Alegre, 984'
         ]

df[ df['LOGRADOURO'].isin(lista)]
```

## Exemplo usando expressão regular para correspondências

```python
import re
df[df.LOGRADOURO.str.contains('MONTE ALEGRE', flags=re.IGNORECASE).fillna(False)][df.BAIRRO.str.contains('PERDIZES', regex= True, na=False,  flags=re.IGNORECASE)]
```
