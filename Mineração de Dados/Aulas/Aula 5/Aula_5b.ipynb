# Usando uma árvore de decisão para regressão

```python
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt


rng = np.random.RandomState(42)
```

## Criando a amostragem

```python
X = np.sort(5 * rng.rand(100, 1), axis=0)
y = np.sin(X).ravel()
```

```python
plt.plot(X,y)
```

## Adicionando ruído aos dados

```python
y[::2] += 0.5 * (0.5 - rng.rand(50))
```

```python
plt.plot(X,y)
```

## Criando a árvore de regressão

```python
regr1 = tree.DecisionTreeRegressor(max_depth=2, random_state=42)
regr1.fit(X, y)
```

```python
regr2 = tree.DecisionTreeRegressor(max_depth=5, random_state=42)
regr2.fit(X, y)
```

```python
regr3 = tree.DecisionTreeRegressor(max_depth=10, random_state=42)
regr3.fit(X, y)
```

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
```

```python
y_1 = regr1.predict(X_test)
y_2 = regr2.predict(X_test)
y_3 = regr3.predict(X_test)
```

## Visualização dos resultados

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, c='k', s=50, label='data')
plt.plot(X_test, y_1, label="max_depth=2", linewidth=5)
plt.plot(X_test, y_2, label="max_depth=5", linewidth=3)
plt.plot(X_test, y_3, label="max_depth=10", linewidth=3)
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
```
