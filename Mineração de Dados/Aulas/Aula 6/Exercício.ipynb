```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
```

```python
df = pd.read_csv('banking.txt')
df
```

```python
df.replace('unknown', np.NaN, inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
dic = {}
for col in df.select_dtypes(object):
    df[col], aux = pd.factorize( df[col] )
    dic[col] = {val: i for i,val in enumerate(aux)}   
```

```python
df.reset_index(inplace=True)
df
```

# Undersampling

```python
classes = [0,1]
```

```python
contagem = df['y'].value_counts().sort_index()
```

```python
contagem.argmin(), contagem.min()
```

```python
dicionario = {i: df[ df['y'] == i] for i in classes}
```

```python
for key in dicionario:
  dicionario[key] = dicionario[key].sample(contagem.min(), replace=True)
```

```python
df2 = pd.concat([dicionario[i] for i in classes], ignore_index=True, sort=False)
df2
```

```python
X = df2.drop(['index', 'y'], axis=1)
X
```

```python
y = df2['y']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Classificador

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=20, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
