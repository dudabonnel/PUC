### Contador binário

```python
def processar(a, N):
    for i in range(N):
        print(a[i], end='')
    print('')

def enumerar(k,a,N):
    if k==N:
        processar(a,N)
        return # encerra a chamada recursiva
    enumerar(k+1,a,N)
    a[k] = 1
    enumerar(k+1,a,N)
    a[k] = 0

def contador(N):
    a = [0 for i in range(N)] # Criar uma lista de zeros
    enumerar(0,a,N)
```

```python
def valorDecimal(a,N):
    soma = 0
    aa = a[::-1]
    for i in range(N):
        soma += aa[i] * 2**i
    return soma

def enumerar(k,a,N):
    if k==N:
        # fazer alguma coisa com essa combinação
        print(a, valorDecimal(a,N)) 
        return
    enumerar(k+1,a,N)
    a[k] = 1
    enumerar(k+1,a,N)
    a[k] = 0

def contador(N):
    a = [0 for i in range(N)] # Criar uma lista de zeros
    enumerar(0,a,N)
```

```python
def seleciona(a,b):
  c = []
  for i,val in enumerate(b):
    if a[i] == 1:
      c.append(val)
  return c

def enumerar(k,a,b,N):
    if k==N:
        # fazer alguma coisa com essa combinação
        print(a, seleciona(a,b)) 
        return
    enumerar(k+1,a,b,N)
    a[k] = 1
    enumerar(k+1,a,b,N)
    a[k] = 0

def subconjuntos(b):
    N = len(b)
    a = [0 for i in range(N)] 
    enumerar(0,a,b,N)
```

### Permutações em strings

```python
def processar(a):
    print(''.join(a))

def trocar(a, i, j):
    t = a[i]
    a[i] = a[j]
    a[j] = t

def enumerar(k,a,n):
    if k==n:
        processar(a)
        return

    for i in range(k,n):
        trocar(a,k,i)
        enumerar(k+1,a,n)
        trocar(a,k,i) # Limpeza

def permutacoes(str):
    a = list(str)
    n = len(a)
    enumerar(0,a,n)
```

### O problema N-Queens

```python
# Testa se há necessidade de Backtracking
def backtrack(k,a,n):
    # Conflito diagonal
    for i in range(k):
        if a[i]-a[k] == k-i or a[k]-a[i] == k-i:
            return True
    return False
            
def processar(a):
    print(a)

def trocar(a, i, j):
    t = a[i]
    a[i] = a[j]
    a[j] = t

def enumerar(k,a,n):
    if k==n:
        processar(a)
        return

    for i in range(k,n):
        trocar(a,k,i)
        if not backtrack(k,a,n):
            enumerar(k+1,a,n)
        trocar(a,k,i) # Limpeza

def Queens(n):
    a = [i for i in range(n)]
    enumerar(0,a,n)
```

### Knight Tour Problem

```python
def conflito(x,y,sol,N):
    # Conflito posição já visitada e/ou movimento inválido
    if x>=0 and x<N and y>=0 and y<N and sol[x][y] == -1:
        return False
    return True
            
def processar(sol,N):
    for i in range(N):
        for j in range(N):
            print(f'{sol[i][j]:02d} ', end='')
        print('')

def enumerar(x,y,iMove,sol,N,movimentos):
    if iMove == N*N:
        processar(sol,N)
        return True

    for mov_x,mov_y in movimentos:
        next_x = x + mov_x
        next_y = y + mov_y

        if not conflito(next_x,next_y,sol,N):
            sol[next_x][next_y] = iMove
            if enumerar(next_x,next_y,iMove+1,sol,N,movimentos):
                return True
            else:
                sol[next_x][next_y] = -1 # Limpeza
    
    return False
    
def KnightTour(N):
    sol = [[-1 for i in range(N)] for j in range(N)]
    sol[0][0] = 0
    movimentos = [(2,1), (1,2), (-1,2), (-2,1), (-2,-1), (-1,-2),  (1,-2),  (2,-1)]

    if enumerar(0,0,1,sol,N,movimentos) == False:
        print('Solução não existe.')
```

### Sudoku

```python
def processar(grid):
    N = len(grid)
    for i in range(N):
        for j in range(N):
            print(f'{grid[i][j]} ', end='')
        print('')

def conflito_linha(grid, row, num):
    N = len(grid[0]) # Quantidade de colunas
    for col in range(N):
        if grid[row][col] == num:
            return True
    return False

def conflito_coluna(grid, col, num):
    N = len(grid) # Quantidade de linhas
    for row in range(N):
        if grid[row][col] == num:
            return True
    return False

def conflito_quadrado(grid,startRow, startCol, num):
    N = int(len(grid)**0.5) # Tamanho quadrado
    for row in range(N):
        for col in range(N):
            if grid[row+startRow][col+startCol] == num:
                return True
    return False

def conflito(grid,row,col,num):
    N = int(len(grid)**0.5) # Tamanho quadrado
    startRow = row - row%N 
    startCol = col - col%N
    if not conflito_linha(grid,row,num) and \
        not conflito_coluna(grid,col,num) and \
        not conflito_quadrado(grid,startRow,startCol,num):
        return False
    return True

def encontraVazio(grid):
    N = len(grid)
    for row in range(N):
        for col in range(N):
            if grid[row][col] == 0:
                return True, (row,col)
    return False,(-1,-1)

def enumerar(grid):

    N = len(grid)

    status, (row,col) = encontraVazio(grid)
    if status == False:
        processar(grid)
        return True

    for num in range(1,N+1):
        
        if not conflito(grid,row,col,num):
            grid[row][col] = num
            if enumerar(grid):
                return True
            grid[row][col] = 0 # Etapa de Backtracking (REMOVE)

    return False

def sudoku(grid):
    if not enumerar(grid):
        print('Sem solução!')  
```

### Labirinto

```python
def imprime(M):
    m,n = len(M), len(M[0])
    for i in range(m):
        for j in range(n):
            if M[i][j] == -2:
                print('███',end='')
            if M[i][j] == -1:
                print('   ',end='')
            if M[i][j] >= 0:
                print(f'{M[i][j]:02d} ', end='')
        print('')
    print('')
    # Aguarda apertar a tecla e limpa a saída
    input('')
    clear_output(wait = True)

def enumerar(M, movimentos, inicio, fim):

    if inicio[0] == fim[0] and inicio[1] == fim[1]:
        imprime(M)
        return M[inicio[0]][inicio[1]]

    # Testa todos os movimentos a partir da posição atual
    for mov in movimentos:
        L = inicio[0] + mov[0]
        C = inicio[1] + mov[1]

        # Verifica se o movimento é válido
        if M[L][C] == -1:
            M[L][C] = M[inicio[0]][inicio[1]] + 1
            imprime(M)
            passos = enumerar(M,movimentos,(L,C),fim)
            if passos > 0:
                return passos
    
    return 0

def labirinto(D, inicio, fim):

    # Cima -> Direita -> Baixo -> Esquerda
    movimentos = [(0,1), (1,0), (0,-1), (-1,0)]

    m,n = len(D), len(D[0])
    M = [[0 for i in range(m)] for j in range(n)]
    for i in range(m):
        for j in range(n):
            if D[i][j] == 1:
                M[i][j] = -2
            else:
                M[i][j] = -1

    M[inicio[0]][inicio[1]] = 0 # Posição inicial
    resposta = enumerar(M,movimentos,inicio,fim) 
    if resposta == 0:
        print('Sem solução')              
    else:
        print(f'Existe uma solução em {resposta} passos.')
        imprime(M)
```
