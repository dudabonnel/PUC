```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from random import random
```

```python
n = 100000

# Distribuição normal
Passo1 = stats.norm.rvs(10,1.5,n)
```

```python
plt.hist( Passo1 , bins=100, edgecolor='black');
```

```python
left = 12
mode = 14
right = 20

c = (mode - left) / (right - left)
loc = left
scale = right - left

Passo2 = stats.triang.rvs(c,loc,scale,n)
```

```python
plt.hist( Passo2 , bins=100, edgecolor='black');
```

```python
Passo3 = np.zeros(n)
for i in range(n):
    if random() <= 0.1:
        # retrabalho 12-14
        Passo3[i] = stats.uniform.rvs(12,2)
    else:
        # normal 8-10
        Passo3[i] = stats.uniform.rvs(8,2)
```

```python
plt.hist( Passo3 , bins=100, edgecolor='black');
```

```python
# Distribuição normal
Passo4 = stats.uniform.rvs(5,8,n)
```

```python
plt.hist( Passo4 , bins=100, edgecolor='black');
```

```python
Tempo = np.zeros(n)
for i in range(n):
    Tempo[i] = Passo1[i] + Passo2[i] + Passo3[i] + Passo4[i]
```

```python
plt.hist( Tempo , bins=100, edgecolor='black');
```

```python
resumo = stats.describe(Tempo)

print('Valor mínimo: %.6f'%resumo.minmax[0])
print('Valor máximo: %.6f'%resumo.minmax[1])
print('Média: %.6f'%resumo.mean)
print('Desvio padrão: %.6f'%np.sqrt(resumo.variance))
```

```python
x = np.linspace(Tempo.min(),Tempo.max(),100)

mu = resumo.mean
sigma = np.sqrt(resumo.variance)

plt.plot( x , stats.norm.pdf(x,mu,sigma) )
plt.axvline(x=45, ymin=0, ymax=1, color = 'r')
```

```python
# c, loc, scale -> c é o parâmetro da distribuição
prob = stats.norm.sf(45,mu,sigma)
print('A probabilidade de um tempor maior que 45 min é %.3f%%'%(prob*100) )
```
