```python
import geopandas as gpd
import folium
```

```python
gdf = gpd.read_file('/content/SP_Distritos.geojson')
gdf
```

```python
gdf['centroid'] = gdf['geometry'].centroid
```

```python
p = (gdf['centroid'][0].y, gdf['centroid'][0].x)
nomes = gdf['ds_nome'].to_list()
```

```python
m = folium.Map(location=p)

for i in range(len(nomes)):
    p = (gdf['centroid'][i].y, gdf['centroid'][i].x)
    folium.Marker(p, popup=nomes[i]).add_to(m)
m
```
