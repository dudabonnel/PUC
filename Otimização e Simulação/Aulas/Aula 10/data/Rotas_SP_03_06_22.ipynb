```python
import osmnx as ox
import networkx as nx
import folium
from osmnx.utils_graph import get_route_edge_attributes
```

```python
G = ox.load_graphml('/content/drive/MyDrive/data/mapas/sp.graphml')
#G = ox.graph_from_place('São Paulo', network_type='drive')
```

```python
origem = 'Autódromo de Interlagos'
O = ox.geocoder.geocode(origem)

#destino = 'Avenida Paulista, 171'
destino = 'Estádio de Itaquera'
D = ox.geocoder.geocode(destino)
```

```python
no_origem, no_destino = ox.nearest_nodes(G, [O[1],D[1]], [O[0],D[0]])
```

## Rota de menor distância

```python
route = nx.shortest_path(G, no_origem, no_destino, weight='length')
#route = nx.shortest_path(G, no_origem, no_destino) # Errado!
```

```python
edge_lengths = get_route_edge_attributes(G,route,'length')
total_route_length = sum(edge_lengths)/1000
print(f'Distância percorrida de {total_route_length:.2f} km')
```

```python
mapa = ox.plot_route_folium(G, route)
folium.Marker(location=O).add_to(mapa)
folium.Marker(location=D).add_to(mapa)
mapa
```
